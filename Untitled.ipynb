```python
d=1
D=0

qs = range(1, 5)
Qs = range(1, 3)

ps = range(1, 5)
Ps = range(3, 5)
size = len(df['ValueLog'])
percentile = int(0.8 * size)

train = df['ValueLog'][:percentile]
test = df['High'][percentile:]
```

```python
from itertools import product
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
#warnings.filterwarnings('ignore')
import statsmodels.api as sm
from tqdm import tqdm


results = []
best_aic = float("inf")

for param in tqdm(parameters_list):
    try:
        model = sm.tsa.statespace.SARIMAX(
            train, 
            order=(param[0], d, param[1]), 
            seasonal_order=(param[2], D, param[3], 12)
        ).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

```python
plt.figure(figsize=(15,15))

plt.subplot(211)
best_model.resid[1:].plot()

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[1:].values.squeeze(), lags=46, ax=ax);
```

```python
train_pred = np.exp(best_model.fittedvalues)

plt.figure(figsize=(15,7))

df['High'].plot()
train_pred.plot(c='r');
```

```python
predict_len = size - percentile
pred = np.exp(best_model.predict(start=percentile, end=percentile + predict_len - 1))

plt.figure(figsize=(15,7))
df['High'].plot()
pred.plot(c='g')
train_pred.plot(c='r');
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mean_absolute_percentage_error(pred, test)
```
